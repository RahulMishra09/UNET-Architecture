```python
import os
import numpy as np
import nibabel as nib
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import warnings
import gc
warnings.filterwarnings('ignore')

# Set device and memory management
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Clear GPU cache if using CUDA
if torch.cuda.is_available():
    torch.cuda.empty_cache()

# Configuration
class Config:
    def __init__(self):
        self.data_path = '/kaggle/input/brats20-dataset-training-validation/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData'
        self.batch_size = 2  # Small batch size for memory efficiency
        self.learning_rate = 1e-4
        self.num_epochs = 75  # Reduced for testing
        self.patch_size = (64, 64, 64)  # Even smaller patches
        self.num_classes = 4  # Background, NCR/NET, ED, ET
        self.in_channels = 4  # T1, T1ce, T2, FLAIR
        self.val_split = 0.2
        self.test_split = 0.1
        self.max_patches_per_case = 8  # Limit patches per case
        self.foreground_ratio = 0.1  # Minimum foreground ratio to keep patch

config = Config()

# Lightweight data preprocessor
class DataPreprocessor:
    def __init__(self):
        self.modalities = ['t1', 't1ce', 't2', 'flair']
    
    def normalize_intensity(self, image):
        """Normalize image intensity using z-score normalization"""
        mask = image > 0
        if np.sum(mask) > 0:
            mean = np.mean(image[mask])
            std = np.std(image[mask])
            if std > 0:
                image[mask] = (image[mask] - mean) / std
        return image.astype(np.float32)  # Use float32 to save memory
    
    def extract_random_patches(self, image, segmentation, patch_size, max_patches=8):
        """Extract random patches with some foreground content"""
        patches_img = []
        patches_seg = []
        
        H, W, D = image.shape[1], image.shape[2], image.shape[3]
        patch_h, patch_w, patch_d = patch_size
        
        # Find foreground regions
        foreground_mask = np.any(segmentation > 0, axis=None)
        if not foreground_mask:
            foreground_coords = [(H//2, W//2, D//2)]
        else:
            foreground_indices = np.where(segmentation > 0)
            if len(foreground_indices[0]) > 0:
                # Sample some foreground coordinates
                n_fg = min(len(foreground_indices[0]), max_patches // 2)
                fg_sample_idx = np.random.choice(len(foreground_indices[0]), n_fg, replace=False)
                foreground_coords = list(zip(
                    foreground_indices[0][fg_sample_idx],
                    foreground_indices[1][fg_sample_idx],
                    foreground_indices[2][fg_sample_idx]
                ))
            else:
                foreground_coords = [(H//2, W//2, D//2)]
        
        # Add some random coordinates
        n_random = max_patches - len(foreground_coords)
        for _ in range(n_random):
            x = np.random.randint(patch_h//2, H - patch_h//2)
            y = np.random.randint(patch_w//2, W - patch_w//2)
            z = np.random.randint(patch_d//2, D - patch_d//2)
            foreground_coords.append((x, y, z))
        
        # Extract patches around these coordinates
        for x, y, z in foreground_coords[:max_patches]:
            x1 = max(0, x - patch_h//2)
            x2 = min(H, x1 + patch_h)
            y1 = max(0, y - patch_w//2)
            y2 = min(W, y1 + patch_w)
            z1 = max(0, z - patch_d//2)
            z2 = min(D, z1 + patch_d)
            
            patch_img = image[:, x1:x2, y1:y2, z1:z2]
            patch_seg = segmentation[x1:x2, y1:y2, z1:z2]
            
            # Pad if necessary
            if patch_img.shape[1:] != tuple(patch_size):
                pad_x = patch_size[0] - patch_img.shape[1]
                pad_y = patch_size[1] - patch_img.shape[2]
                pad_z = patch_size[2] - patch_img.shape[3]
                
                patch_img = np.pad(patch_img, 
                                 ((0, 0), (0, pad_x), (0, pad_y), (0, pad_z)), 
                                 mode='constant')
                patch_seg = np.pad(patch_seg, 
                                 ((0, pad_x), (0, pad_y), (0, pad_z)), 
                                 mode='constant')
            
            # Check if patch has enough foreground
            fg_ratio = np.sum(patch_seg > 0) / patch_seg.size
            if fg_ratio >= config.foreground_ratio or len(patches_img) < 2:  # Keep at least 2 patches
                patches_img.append(patch_img.astype(np.float32))
                patches_seg.append(patch_seg.astype(np.int64))
            
            if len(patches_img) >= max_patches:
                break
        
        return patches_img, patches_seg
    
    def load_nifti(self, filepath):
        """Load NIfTI file and return image data"""
        nii = nib.load(filepath)
        data = nii.get_fdata().astype(np.float32)
        return data
    
    def preprocess_case(self, case_folder):
        """Preprocess a single case and extract patches"""
        case_name = os.path.basename(case_folder)
        
        try:
            # Load all modalities
            images = []
            for modality in self.modalities:
                filepath = os.path.join(case_folder, f"{case_name}_{modality}.nii")
                if os.path.exists(filepath):
                    image = self.load_nifti(filepath)
                    image = self.normalize_intensity(image)
                    images.append(image)
                else:
                    print(f"Warning: {filepath} not found")
                    return None, None
            
            # Stack modalities
            image = np.stack(images, axis=0)  # Shape: (4, H, W, D)
            
            # Load segmentation mask
            seg_filepath = os.path.join(case_folder, f"{case_name}_seg.nii")
            if os.path.exists(seg_filepath):
                segmentation = self.load_nifti(seg_filepath)
                
                # Convert BraTS labels to our class indices
                seg_processed = np.zeros_like(segmentation, dtype=np.int64)
                seg_processed[segmentation == 1] = 1  # NCR/NET
                seg_processed[segmentation == 2] = 2  # ED
                seg_processed[segmentation == 4] = 3  # ET
                
                # Extract patches
                patches_img, patches_seg = self.extract_random_patches(
                    image, seg_processed, config.patch_size, config.max_patches_per_case
                )
                
                # Clean up memory
                del image, segmentation, seg_processed
                gc.collect()
                
                return patches_img, patches_seg
            else:
                print(f"Warning: {seg_filepath} not found")
                return None, None
                
        except Exception as e:
            print(f"Error processing {case_folder}: {e}")
            return None, None

# Memory-efficient dataset
class BraTSPatchDataset(Dataset):
    def __init__(self, case_folders, preprocessor, max_cases=None):
        self.case_folders = case_folders
        self.preprocessor = preprocessor
        self.patches = []
        self.segmentations = []
        
        # Limit number of cases for memory efficiency
        if max_cases:
            case_folders = case_folders[:max_cases]
        
        print(f"Processing {len(case_folders)} cases...")
        valid_cases = 0
        
        for i, case_folder in enumerate(tqdm(case_folders)):
            if self._validate_case(case_folder):
                patches_img, patches_seg = self.preprocessor.preprocess_case(case_folder)
                if patches_img is not None and len(patches_img) > 0:
                    self.patches.extend(patches_img)
                    self.segmentations.extend(patches_seg)
                    valid_cases += 1
                    
                    # Periodically clean memory
                    if i % 10 == 0:
                        gc.collect()
                        if torch.cuda.is_available():
                            torch.cuda.empty_cache()
        
        print(f"Loaded {len(self.patches)} patches from {valid_cases} valid cases")
    
    def _validate_case(self, case_folder):
        """Check if case has all required files"""
        case_name = os.path.basename(case_folder)
        required_files = [f"{case_name}_{mod}.nii" for mod in self.preprocessor.modalities]
        required_files.append(f"{case_name}_seg.nii")
        
        for file in required_files:
            if not os.path.exists(os.path.join(case_folder, file)):
                return False
        return True
    
    def __len__(self):
        return len(self.patches)
    
    def __getitem__(self, idx):
        image = torch.FloatTensor(self.patches[idx])
        segmentation = torch.LongTensor(self.segmentations[idx])
        return image, segmentation

# Lightweight 3D U-Net
class Conv3DBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, padding=1):
        super(Conv3DBlock, self).__init__()
        self.conv1 = nn.Conv3d(in_channels, out_channels, kernel_size, padding=padding, bias=False)
        self.bn1 = nn.BatchNorm3d(out_channels)
        self.conv2 = nn.Conv3d(out_channels, out_channels, kernel_size, padding=padding, bias=False)
        self.bn2 = nn.BatchNorm3d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.dropout = nn.Dropout3d(0.1)
    
    def forward(self, x):
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.dropout(x)
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class LightweightUNet3D(nn.Module):
    def __init__(self, in_channels, num_classes):
        super(LightweightUNet3D, self).__init__()
        
        # Encoder (very lightweight)
        self.enc1 = Conv3DBlock(in_channels, 16)
        self.enc2 = Conv3DBlock(16, 32)
        self.enc3 = Conv3DBlock(32, 64)
        
        # Bottleneck
        self.bottleneck = Conv3DBlock(64, 128)
        
        # Decoder
        self.upconv3 = nn.ConvTranspose3d(128, 64, 2, stride=2)
        self.dec3 = Conv3DBlock(128, 64)
        
        self.upconv2 = nn.ConvTranspose3d(64, 32, 2, stride=2)
        self.dec2 = Conv3DBlock(64, 32)
        
        self.upconv1 = nn.ConvTranspose3d(32, 16, 2, stride=2)
        self.dec1 = Conv3DBlock(32, 16)
        
        # Final layer
        self.final_conv = nn.Conv3d(16, num_classes, 1)
        
        # Pooling
        self.pool = nn.MaxPool3d(2)
        
    def forward(self, x):
        # Encoder
        enc1 = self.enc1(x)
        enc2 = self.enc2(self.pool(enc1))
        enc3 = self.enc3(self.pool(enc2))
        
        # Bottleneck
        bottleneck = self.bottleneck(self.pool(enc3))
        
        # Decoder
        dec3 = self.upconv3(bottleneck)
        dec3 = torch.cat([dec3, enc3], dim=1)
        dec3 = self.dec3(dec3)
        
        dec2 = self.upconv2(dec3)
        dec2 = torch.cat([dec2, enc2], dim=1)
        dec2 = self.dec2(dec2)
        
        dec1 = self.upconv1(dec2)
        dec1 = torch.cat([dec1, enc1], dim=1)
        dec1 = self.dec1(dec1)
        
        # Final prediction
        output = self.final_conv(dec1)
        
        return output

# Simple loss functions
class DiceLoss(nn.Module):
    def __init__(self, smooth=1e-5):
        super(DiceLoss, self).__init__()
        self.smooth = smooth
    
    def forward(self, predictions, targets):
        predictions = F.softmax(predictions, dim=1)
        dice_loss = 0
        
        for class_idx in range(predictions.shape[1]):
            pred_class = predictions[:, class_idx]
            target_class = (targets == class_idx).float()
            
            intersection = (pred_class * target_class).sum()
            union = pred_class.sum() + target_class.sum()
            
            dice = (2 * intersection + self.smooth) / (union + self.smooth)
            dice_loss += 1 - dice
        
        return dice_loss / predictions.shape[1]

# Simplified metrics calculator
class MetricsCalculator:
    def __init__(self, num_classes):
        self.num_classes = num_classes
    
    def dice_score(self, pred, target, class_idx):
        """Calculate Dice score for a specific class"""
        pred_class = (pred == class_idx).float()
        target_class = (target == class_idx).float()
        
        intersection = (pred_class * target_class).sum()
        union = pred_class.sum() + target_class.sum()
        
        if union == 0:
            return torch.tensor(1.0 if intersection == 0 else 0.0, device=pred.device)
        
        return (2 * intersection) / union
    
    def calculate_metrics(self, predictions, targets):
        """Calculate all metrics"""
        predictions = torch.argmax(predictions, dim=1)
        
        metrics = {}
        dice_scores = []
        
        for class_idx in range(self.num_classes):
            dice = self.dice_score(predictions, targets, class_idx)
            dice_scores.append(dice.item())
            metrics[f'dice_class_{class_idx}'] = dice.item()
        
        metrics['mean_dice'] = np.mean(dice_scores)
        
        # BraTS specific metrics
        # Whole tumor (WT) = ED + NET + ET (classes 1, 2, 3)
        wt_pred = ((predictions == 1) | (predictions == 2) | (predictions == 3)).float()
        wt_target = ((targets == 1) | (targets == 2) | (targets == 3)).float()
        wt_dice = self.dice_score_binary(wt_pred, wt_target)
        metrics['dice_wt'] = wt_dice.item()
        
        # Tumor core (TC) = NET + ET (classes 1, 3)
        tc_pred = ((predictions == 1) | (predictions == 3)).float()
        tc_target = ((targets == 1) | (targets == 3)).float()
        tc_dice = self.dice_score_binary(tc_pred, tc_target)
        metrics['dice_tc'] = tc_dice.item()
        
        # Enhancing tumor (ET) = ET (class 3)
        et_pred = (predictions == 3).float()
        et_target = (targets == 3).float()
        et_dice = self.dice_score_binary(et_pred, et_target)
        metrics['dice_et'] = et_dice.item()
        
        return metrics
    
    def dice_score_binary(self, pred, target):
        """Calculate Dice score for binary masks"""
        intersection = (pred * target).sum()
        union = pred.sum() + target.sum()
        
        if union == 0:
            return torch.tensor(1.0, device=pred.device)
        
        return (2 * intersection) / union

# Memory-efficient training function
def train_model(model, train_loader, val_loader, num_epochs, device):
    model = model.to(device)
    criterion = DiceLoss()  # Use only Dice loss for simplicity
    optimizer = Adam(model.parameters(), lr=config.learning_rate, weight_decay=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.7, patience=5, verbose=True)
    
    metrics_calculator = MetricsCalculator(config.num_classes)
    
    train_losses = []
    val_losses = []
    val_dice_scores = []
    
    best_dice = 0.0
    
    for epoch in range(num_epochs):
        # Training
        model.train()
        train_loss = 0.0
        train_batches = 0
        
        for batch_idx, (images, masks) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Training")):
            images, masks = images.to(device), masks.to(device)
            
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, masks)
            loss.backward()
            
            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            
            train_loss += loss.item()
            train_batches += 1
            
            if batch_idx % 50 == 0:
                print(f"Batch {batch_idx}, Loss: {loss.item():.4f}")
            
            # Memory cleanup
            del outputs, loss
            if batch_idx % 20 == 0:
                torch.cuda.empty_cache()
        
        avg_train_loss = train_loss / train_batches
        train_losses.append(avg_train_loss)
        
        # Validation
        model.eval()
        val_loss = 0.0
        val_batches = 0
        all_metrics = []
        
        with torch.no_grad():
            for images, masks in tqdm(val_loader, desc="Validation"):
                images, masks = images.to(device), masks.to(device)
                
                outputs = model(images)
                loss = criterion(outputs, masks)
                val_loss += loss.item()
                val_batches += 1
                
                # Calculate metrics
                metrics = metrics_calculator.calculate_metrics(outputs, masks)
                all_metrics.append(metrics)
                
                # Clean up
                del outputs, loss
        
        # Clear cache after validation
        torch.cuda.empty_cache()
        
        avg_val_loss = val_loss / val_batches
        val_losses.append(avg_val_loss)
        
        # Average metrics
        if all_metrics:
            avg_metrics = {}
            for key in all_metrics[0].keys():
                avg_metrics[key] = np.mean([m[key] for m in all_metrics])
            
            val_dice_scores.append(avg_metrics['mean_dice'])
            
            # Update learning rate
            scheduler.step(avg_val_loss)
            
            # Print results
            print(f"Epoch {epoch+1}/{num_epochs}")
            print(f"Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")
            print(f"Mean Dice: {avg_metrics['mean_dice']:.4f}")
            print(f"WT Dice: {avg_metrics['dice_wt']:.4f}, TC Dice: {avg_metrics['dice_tc']:.4f}, ET Dice: {avg_metrics['dice_et']:.4f}")
            print(f"Learning Rate: {optimizer.param_groups[0]['lr']:.6f}")
            print("-" * 50)
            
            # Save best model
            if avg_metrics['mean_dice'] > best_dice:
                best_dice = avg_metrics['mean_dice']
                torch.save(model.state_dict(), 'best_model.pth')
                print(f"New best model saved with Dice: {best_dice:.4f}")
    
    return train_losses, val_losses, val_dice_scores

# Main execution with memory management
def main():
    # Initialize preprocessor
    preprocessor = DataPreprocessor()
    
    # Get limited number of case folders for testing
    all_case_folders = [os.path.join(config.data_path, folder) 
                       for folder in os.listdir(config.data_path) 
                       if os.path.isdir(os.path.join(config.data_path, folder))]
    
    # Use subset for testing
    case_folders = all_case_folders[:300]  # Use only first 50 cases
    print(f"Using {len(case_folders)} case folders for testing")
    
    # Split data
    train_val_folders, test_folders = train_test_split(
        case_folders, test_size=config.test_split, random_state=42
    )
    
    train_folders, val_folders = train_test_split(
        train_val_folders, test_size=config.val_split, random_state=42
    )
    
    print(f"Train: {len(train_folders)}, Val: {len(val_folders)}, Test: {len(test_folders)}")
    
    # Create datasets
    train_dataset = BraTSPatchDataset(train_folders, preprocessor)
    val_dataset = BraTSPatchDataset(val_folders, preprocessor)
    
    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=config.batch_size, shuffle=True, 
                             num_workers=0, pin_memory=False)  # Reduced workers
    val_loader = DataLoader(val_dataset, batch_size=config.batch_size, shuffle=False, 
                           num_workers=0, pin_memory=False)
    
    # Create lightweight model
    model = LightweightUNet3D(config.in_channels, config.num_classes)
    print(f"Model created with {sum(p.numel() for p in model.parameters())} parameters")
    
    # Train model
    print("Starting training...")
    train_losses, val_losses, val_dice_scores = train_model(
        model, train_loader, val_loader, config.num_epochs, device
    )
    
    # Plot results
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(val_dice_scores, label='Val Dice Score')
    plt.title('Validation Dice Score')
    plt.xlabel('Epoch')
    plt.ylabel('Dice Score')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('training_results.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print(f"Training completed. Best validation Dice: {max(val_dice_scores):.4f}")

if __name__ == "__main__":
    main()
```
